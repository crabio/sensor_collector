```python
import pandas as pd
import plotly.express as px

df = pd.read_csv('sensor-data-2024-07-24-20-12-28.310334.csv.gz')
df['dt'] = pd.to_datetime(df['dt'],unit='ms')
df['user_acc_dt_diff'] = (df.loc[df['user_acc_x'].notna(), 'dt'] - df.loc[df['user_acc_x'].notna(), 'dt'].shift().fillna(pd.NaT)).dt.total_seconds() * 1000
df['acc_dt_diff'] = (df.loc[df['acc_x'].notna(), 'dt'] - df.loc[df['acc_x'].notna(), 'dt'].shift().fillna(pd.NaT)).dt.total_seconds() * 1000
df['gyr_dt_diff'] = (df.loc[df['gyr_x'].notna(), 'dt'] - df.loc[df['gyr_x'].notna(), 'dt'].shift().fillna(pd.NaT)).dt.total_seconds() * 1000
df['mag_dt_diff'] = (df.loc[df['mag_x'].notna(), 'dt'] - df.loc[df['mag_x'].notna(), 'dt'].shift().fillna(pd.NaT)).dt.total_seconds() * 1000
df
```

```python
fig = px.line(df, x='dt', y=['user_acc_dt_diff','acc_dt_diff','gyr_dt_diff','mag_dt_diff'])
fig.update_traces(connectgaps=True)
fig.show()
```

```python
fig = px.line(df, x='dt', y=['acc_x', 'acc_y', 'acc_z'])
fig.update_traces(connectgaps=True)
fig.show()
```

```python
fig = px.line(df, x='dt', y=['mag_x', 'mag_y', 'mag_z'])
fig.update_traces(connectgaps=True)
fig.show()
```
